```python
set1 = {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "apple"}

set3 = set1.intersection(set2)
print(set3)
```

```python
set1 = {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "apple"}

set3 = set1.difference(set2)

print(set3)
```

```python
set1 = {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "apple"}

set3 = set1.symmetric_difference(set2)

print(set3)
```

```python
# Changing the Values using two different methods

thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
print("Value before the updataion :",thisdict)
thisdict["year"] = 2018
print("Value after the updataion :",thisdict)

thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
print("Value before the updataion :",thisdict)
thisdict.update({"year": 2020})
print("Value after the updataion :",thisdict)
```
