```python
# Search for a number X in the tuple [1,4,9,16,25,36,49,64,81,100]..
t = (1, 4, 9, 16, 25, 36, 49, 64, 81, 100)
x=int(input("Enter the number to Search:"))
if x in t:
  print(x,"is Present in",t)
else :
  print(x,"is not present in",t)
```

```python
t = (1, 4, 9, 16, 25, 36, 49, 64, 81, 100)
X = 36

for i in t:
    if i == X:
        print("X is present in the tuple")
        break
else:
    print("X is not present in the tuple")
```

```python
t = (1, 4, 9, 16, 25, 36, 49, 64, 81, 100)
X = 36

for i in t:
    if i != X:
        continue   # skip all elements that are not X
    print("X is present in the tuple")
    break
else:
    print("X is not present in the tuple")
```
