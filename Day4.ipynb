```python
# Remove() - removes the first occurance of any object..

thislist = ["apple", "banana", "cherry", "banana", "kiwi"]
thislist.remove("banana")
print(thislist)
```

```python
#Pop is used to delete the specific element from the desired position

thislist = ["apple", "banana", "cherry"]
thislist.pop(1)
print(thislist)
```

```python
# pop function is used to delete the last element from the list when no parameter is passed
thislist = ["apple", "banana", "cherry"]
thislist.pop()
print(thislist)
```

```python
num=[54,65,42,48,85,99]
num.sort()
print("The Ascending order is:",num)

#Reverse Sorting (descending
num.sort(reverse=True)
print("The Descending order is:",num)
```

```python
 # the sort() method is case sensitive, resulting in all capital letters being sorted before lower case letters:
thislist = ["banana", "Orange", "Kiwi", "cherry"]
thislist.sort()
print(thislist)
```

```python
# reversing the List
thislist.reverse()
print(thislist)
```
